```python
from random import randint

print("Let's begin!Guess the number\n")
# Defining no. of games user wants to play
while True:
    try:
        gamesNum = int(input("No of games you want to play: "))
    except:
        print("Hey!!I do not understand you.Please enter some numbers like 1,2,3,4...\n")
        continue
    break

num = 1 # Counter for no.of games

# This while loop takes care of the total number of games
while num <= gamesNum:
    compGuess = randint(1,25) # Random number genaration
    count = 0
#Asking for user inputs 
    print("Hey guess a number between 1 to 25\n")
    while True:
        try:
            playerGuess = int(input("\nGuess the number:  "))
        except:
            print("Sorry!!I can only compare numbers\n")
            continue    
        if playerGuess == compGuess:
            count += 1
            print(f"\nBingo!! you read my mind in {count} attempts\n")
            break
        elif playerGuess > compGuess:
            print("\nYou are guesing a greater number!")
            if playerGuess > 25:
                print("Hint:I am guessing till 25\n")
            count += 1
            continue
        else:
            print("\nYou are on the smaller side!")
            if playerGuess < 1:
                print("Hint:I am guessing a Non-Zero positive number\n")
            count += 1
            continue

    num += 1
```
